```python
import pandas as pd
```

```python
dados = pd.read_json('/content/drive/MyDrive/Arquivos Pandas: transformação e manipulação de dados/dados_locacao_imoveis.json')
dados.head()
```

```python
dados = pd.json_normalize(dados['dados_locacao'])
dados
```

```python
colunas = list(dados.columns)
colunas
```

```python
dados = dados.explode(colunas[1:])
```

```python
dados.reset_index(drop = True, inplace = True)
```

```python
dados
```

```python
dados['valor_aluguel'] = dados['valor_aluguel'].apply(lambda x : x.replace('$', '').replace(',', '.').strip())
```

```python
dados['valor_aluguel'] = dados['valor_aluguel'].apply(lambda x : x.replace('reais','').strip())
```

```python
import numpy as np
```

```python
dados['valor_aluguel'] = dados['valor_aluguel'].astype(np.float64)
```

```python
dados.info()
```

```python
dados['apartamento'] = dados['apartamento'].str.replace(' \(blocoAP\)', '', regex = True)
```

```python
dados.head()
```

```python
dados['datas_de_pagamento'] = pd.to_datetime(dados['datas_de_pagamento'], format = '%d/%m/%Y')
```

```python
dados['datas_combinadas_pagamento'] = pd.to_datetime(dados['datas_combinadas_pagamento'], format = '%d/%m/%Y')
```

```python
dados.info()
```

```python
dados['atraso'] = (dados['datas_de_pagamento'] - dados['datas_combinadas_pagamento']).dt.days
```

```python
media_atraso = dados.groupby(dados['apartamento'])['atraso'].mean()
```
